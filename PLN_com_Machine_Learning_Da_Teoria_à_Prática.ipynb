### **TIPOS DE APRENDIZAGEM DE MÁQUINA**

Dentro do aprendizado de máquina, existem diferentes formas de treinar algoritmos para que eles gerem um modelo. Entre os principais tipos, podemos destacar:

- **Aprendizado Supervisionado**: Nesse tipo de aprendizado, fornecemos ao algoritmo um conjunto de dados conhecidos. Com esses dados, ele aprende a reconhecer padrões importantes e, posteriormente, consegue identificá-los em outros conjuntos de dados, conhecidos ou não.

- **Aprendizado Não Supervisionado**: Nesse caso, o próprio algoritmo busca padrões e estruturas nos dados, sem receber informações prévias sobre respostas corretas. Por exemplo, ele pode agrupar dados semelhantes em categorias ou identificar tendências.

- **Aprendizado por Reforço (ou por recompensa)**: Aqui, o modelo aprende por meio de recompensas ou penalidades. Ou seja, o algoritmo realiza ações e recebe um “feedback” indicando se a ação foi correta ou não, ajustando seu comportamento para maximizar as recompensas ao longo do tempo.



--------------------------------------------------------------------------------
### **TREINAMENTO DE MODELOS**

Um ***modelo*** é o resultado obtido após o processo de treinamento de um algoritmo de Machine Learning. Ou seja, durante o treinamento, estamos ensinando o algoritmo a aprender padrões, e o que obtemos ao final desse processo é o modelo.

Ao treinarmos um algoritmo com o objetivo de obter um modelo, trabalhamos com base na ***generalização***, que nada mais é do que a capacidade do modelo de reconhecer padrões fora do conjunto de dados de treinamento. Em outras palavras, após o treinamento, devemos estar atentos a certas situações para garantir que o modelo funcione bem com dados novos.

O ***overfitting*** ocorre quando o modelo aprende demais os padrões de um conjunto específico de dados, perdendo a capacidade de reconhecer padrões em outros conjuntos, mesmo que sejam muito semelhantes.

O ***underfitting*** é o oposto: acontece quando o modelo aprende muito pouco, não conseguindo identificar padrões nem no conjunto de dados de treinamento nem em novos dados.

--------------------------------------------------------------------------------
### **CONJUNTO DE DADOS**

Ao treinarmos um algoritmo, utilizamos conjuntos de dados, ou seja, coleções de informações que servem para ensinar o modelo a reconhecer padrões. Normalmente, esses conjuntos são divididos em três partes principais


1.   Conjunto de Treinamento (Training Set): usado para ensinar o modelo a
identificar padrões.
2.   Conjunto de Validação (Validation Set): usado para ajustar parâmetros do modelo e evitar overfitting.
3.   Conjunto de Teste (Test Set): usado para avaliar a capacidade do modelo de generalizar e fazer previsões em dados que ele nunca viu.

--------------------------------------------------------------------------------
### **OUTROS CONCEITOS IMPORTANTES**

**-Engenharia de Recursos** (Feature Engineering): processo que transforma dados brutos em dados mais refinados, facilitando o treinamento do modelo e ajudando-o a aprender padrões de forma mais eficiente.

**-Ajuste de Hiperparâmetros** (Hyperparameter Tuning): consiste em escolher os parâmetros do modelo que não são aprendidos automaticamente durante o treinamento, com o objetivo de melhorar o desempenho e evitar overfitting.

**-Métricas** (Metrics): indicadores utilizados para avaliar o desempenho do modelo, como acurácia, precisão, recall e F1-score, dependendo do tipo de problema.

**-Parada Antecipada** (Early Stopping): técnica que interrompe o treinamento do modelo antes que ele se ajuste demais aos dados de treinamento, prevenindo overfitting e melhorando a capacidade de generalização.

### **NORMALIZAÇÃO**

Desde o início, é importante frisarmos que nem sempre os textos estarão formatados e “bonitinhos”. Às vezes, eles chegarão ao algoritmo desorganizados, com letras maiúsculas e minúsculas misturadas, acentuação inconsistente e cheios de caracteres especiais. Isso pode comprometer a análise e o desempenho de modelos de processamento de linguagem natural. Por isso, a normalização do texto é uma etapa fundamental no pré-processamento, pois garante maior consistência nos dados e facilita o trabalho dos algoritmos.

Agora que possuímos a biblioteca, o primeiro passo para normalizarmos um texto seria converter todos os seus caracteres para letras minúsculas. Para isso, utilizaremos a função *lower()*.

```python
import nltk

texto = "HoJe, *9* eM cAdA 10 bA/nCoS jÁ pOsSuEm uMa pLaTaFoRmA dIg-ItAl, tOrNaNdO rArO eNcOnT*rAr uM qUe oPeRe aPeNaS dE fOrMa fÍsIcA"

##Imprimindo texto original
print("Texto original: " + texto)

##Convertendo o texto para minusculo
texto_minusculo = texto.lower()

##Imprimindo o texto minusculo
print("Texto Minusculo: " + texto_minusculo)
```

Para o próximo passo, iremos remover os caracteres especiais, que podem atrapalhar nosso algoritmo no processamento do texto. Para isso, usaremos [expressões regulares](https://www.devmedia.com.br/iniciando-expressoes-regulares/6557), para identificar esses caracteres indesejados e eliminá-los do texto de forma eficiente.

Importaremos a biblioteca *re*, que é uma biblioteca padrão do Python para trabalharmos com expressões regulares.

```python
import re  # Importa o módulo 're' para usar expressões regulares
```

Após isso, iremos definir uma expressão regular que desempenhará o papel de remover esses elementos indesejados do nosso texto.

```python
import nltk
import re

texto = "HoJe, *9* eM cAdA 10 bA/nCoS jÁ pOsSuEm uMa pLaTaFoRmA dIg-ItAl, tOrNaNdO rArO eNcOnT*rAr uM qUe oPeRe aPeNaS dE fOrMa fÍsIcA"

texto_minusculo = texto.lower()

# Remove todos os caracteres que não são letras, números ou espaços do texto já em minúsculas
# A expressão correta é [^\w\s], com \w para letras/números e \s para espaços
textoLimpo = re.sub(r'[^\w\s]', "", texto_minusculo)

# Exibe o texto convertido para minúsculas e sem caracteres especiais
print("Texto minúsculo e sem caracteres especiais: " + textoLimpo)
```

### **TOKENIZAÇÃO**

A tokenização é o processo de dividir um texto em partes menores e mais fáceis de serem analisadas. Essas partes são chamadas de tokens.Para começar a tokenizar textos, iremos importar duas funções da biblioteca *nltk*:

- *word_tokenize*: divide o texto em palavras, pontuações e números — cada token é uma palavra individual, um caractere de pontuação ou um número.

- *sent_tokenize:* divide o texto em frases — cada token corresponde a uma frase completa extraída do texto.

Para dar eficiência às funções de tokenização, também precisamos baixar um recurso que permite realizar essa separação de forma correta. Esse recurso é conhecido como punkt_t, e ele é responsável por identificar automaticamente os limites de palavras e frases com base em regras e estatísticas da linguagem. Em abordagens mais modernas, essa função é realizada por um Tokenizador especializado (usando técnicas como Byte-Pair Encoding (BPE) ou WordPiece) que divide o texto em subunidades para alimentar modelos avançados.

```python
import nltk

#Baixando o modelo
nltk.download('punkt_tab')

#Importando as funçãoes de tokenização da biblioteca nltk
from nltk.tokenize import word_tokenize, sent_tokenize
```

Para testar as funções de tokenização, utilizaremos como exemplo a seguinte frase: 'Os bancos digitais estão revolucionando a forma como lidamos com o dinheiro no cotidiano.'

```python
import nltk
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize, sent_tokenize

texto = "Os bancos digitais estão revolucionando a forma como lidamos com o dinheiro no cotidiano."

#Tokenizando o texto em palavras
palavras = nltk.word_tokenize(texto)

#Tokenizando o texto em frases
frases = nltk.sent_tokenize(texto)

#Imprimindo o texto tokenizado em palavras
print("Texto tokenizado em palavras: " + str(palavras))

#Imprimindo o texto tokenizado em frases
print("Texto tokenizado em frases: " + str(frases))
```

Observe como essas funções dividirão nosso texto em partes menores e mais simples de serem analisadas. Um texto pesado virou algo estruturado, onde conseguimos identificar palavras e frases individualmente.

### **REMOÇÃO DE STOPWORDS**

***Stopwords*** são palavras que aparecem com muita frequência em um texto, mas que geralmente não contribuem de forma significativa para o processamento ou classificação desse texto. Elas têm a função de dar fluidez e atender às normas da língua, mas não alteram o conteúdo essencial.

Exemplos comuns são artigos como o, a, os, as. Essas palavras ajudam na construção de frases mais naturais, mas pouco influenciam quando o objetivo é identificar o tema de um texto.

Por exemplo, na frase:
"O Python é uma linguagem de programação",
as palavras-chave realmente relevantes para classificação são Python, linguagem e programação.

Por isso é de suma importancia, remove-las, pois ao remover as ***stopwords***, o algoritmo consegue focar nos termos que indicam sobre o que o texto realmente trata, e parar de gastar processamento com palavras que pouco influenciam no resultado final.

A biblioteca *NLTK* fornece suporte para a remoção de stopwords em 16 idiomas diferentes, incluindo o português. Para isso, utilizamos o corpus, que contém uma lista pré-definida dessas palavras para cada língua. No caso do exemplo anterior, podemos combinar o uso do corpus de stopwords com a função word_tokenize para separar o texto em tokens e, em seguida, eliminar as palavras irrelevantes.

Primeiro vamos importar e baixar o necessário:

```python
import nltk
#Importando o corpus
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

#Baixando o pacote de stopwords
nltk.download('stopwords')
nltk.download('punkt_tab')
```

Após configurar o uso das stopwords, definiremos um texto de exemplo e aplicaremos a função word_tokenize para tokenizá-lo

Agora que já tokenizamos o nosso texto, o próximo passo é remover as stopwords. Para isso, vamos criar uma variável que vai armazenar a lista de stopwords em português, passando o parâmetro "portuguese".

Com essa lista, conseguimos comparar cada token do texto e remover as palavras irrelevantes — mantendo apenas aquelas que realmente contribuem para a nossa análise.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize


texto = "Nos ultimos tempos o Python se tornou uma das linguagens mais populares do mundo, sendo amplamente utilizada tanto em ambientes acadêmicos quanto no mercado de trabalho."
textoTokenizado = nltk.word_tokenize(texto)

# Carregando as stopwords da língua portuguesa e armazenando na variável ptStopwords
ptStopwords = set(stopwords.words('portuguese'))

#imprimindo a lista de stopwords
print(" - " + str(ptStopwords))
```

Feito isso, vamos criar uma função que vai verificar se cada token do texto tokenizado está presente na nossa lista de stopwords em português (ptStopwords).

Essa função vai retornar o texto já filtrado — ou seja, sem as stopwords — mantendo apenas as palavras que realmente carregam significado para a nossa análise.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize


texto = "Nos ultimos tempos o Python se tornou uma das linguagens mais populares do mundo, sendo amplamente utilizada tanto em ambientes acadêmicos quanto no mercado de trabalho."
textoTokenizado = nltk.word_tokenize(texto)


ptStopwords = set(stopwords.words('portuguese'))

# Definindo a função para remover stopwords
# Parâmetros:
# - textoTokenizado: lista de palavras do texto já tokenizado
# - ptStopwords: conjunto de stopwords em português
# Retorno:
# - lista de palavras do texto sem as stopwords
def removeStopwords(textoTokenizado, ptStopwords):
    # Mantém apenas as palavras que NÃO estão na lista de stopwords
    textoSemStopwords = [palavra for palavra in textoTokenizado if palavra.lower() not in ptStopwords]
    return textoSemStopwords

#Chamando a função
textoSemStopwords = removeStopwords(textoTokenizado, ptStopwords)

#Imprimindo o texto sem stopwords
print("Texto sem stopwords: " + str(textoSemStopwords))

```

#### **REMOVENDO STOPWODS DE MANEIRA PERSONALIZADA**

Por mais que já existam bibliotecas, funções ou modelos prontos para a remoção de stopwords, nem sempre esse material é o mais adequado para o que estamos trabalhando.

Em algumas situações, palavras removidas por essas ferramentas — por serem classificadas como stopwords — podem, na verdade, ter grande importância para o nosso processamento ou análise.

Sendo assim, vamos criar dois projetos que permitam a personalização da nossa própria biblioteca de stopwords.

1. Algoritimo que permite 100% de personalização das stopwods:

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')  # Só se ainda não baixou o tokenizer

texto = "Nos últimos tempos o Python se tornou uma das linguagens mais populares do mundo, sendo amplamente utilizada tanto em ambientes acadêmicos quanto no mercado de trabalho."

# Tokeniza o texto
textoTokenizado = word_tokenize(texto)

# Defina sua própria lista personalizada de stopwords (sem usar a base do NLTK)
stopwords_customizadas = set([
    "o", "a", "as", "os", "de", "do", "da", "dos", "das", "e", "é",
    "em", "um", "uma", "uns", "umas", "que", "com", "por", "para",
    "nos", "no", "na", "nos", "nas", "tanto", "como", "seja",
    "sendo", "dos", "das", "tempos"  # Exemplo de palavra personalizada
])

# Função para remover as stopwords definidas
def removeStopwords(textoTokenizado, stopwords_set):
    return [palavra for palavra in textoTokenizado if palavra.lower() not in stopwords_set]

# Aplicar remoção de stopwords
textoSemStopwords = removeStopwords(textoTokenizado, stopwords_customizadas)

print("Texto original tokenizado:")
print(textoTokenizado)

print("\nTexto após remoção das stopwords personalizadas:")
print(textoSemStopwords)

```

2. Algoritmo que utiliza a base de stopwords em português da biblioteca NLTK, porém permite a inclusão de novas palavras e a remoção de palavras existentes na lista carregada:

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('stopwords')

# Texto de exemplo
texto = "Nos últimos tempos o Python se tornou uma das linguagens mais populares do mundo, sendo amplamente utilizada tanto em ambientes acadêmicos quanto no mercado de trabalho."

# Tokenizando o texto (separando em palavras)
textoTokenizado = word_tokenize(texto)

# Carregando as stopwords padrão do NLTK para português
stopwords_base = set(stopwords.words('portuguese'))

# Função para criar uma lista personalizada de stopwords,
# permitindo adicionar novas palavras à lista base
def personalizar_stopwords(palavras_para_adicionar=None):
    # Copia as stopwords base para evitar alterar o original
    stopwords_personalizadas = set(stopwords_base)

    # Se houver palavras para adicionar, insere na lista personalizada
    if palavras_para_adicionar:
        for palavra in palavras_para_adicionar:
            stopwords_personalizadas.add(palavra.lower())

    return stopwords_personalizadas

# Criando a lista de stopwords personalizada, adicionando "tempos"
stopwords_customizadas = personalizar_stopwords(
    palavras_para_adicionar=["tempos"]
)

# Função para remover stopwords do texto tokenizado
def removeStopwords(textoTokenizado, stopwords_set):
    # Retorna somente as palavras que NÃO estão nas stopwords
    return [palavra for palavra in textoTokenizado if palavra.lower() not in stopwords_set]

# Aplicando a remoção de stopwords personalizada
textoSemStopwords = removeStopwords(textoTokenizado, stopwords_customizadas)

# Mostrando os resultados
print("Texto original tokenizado:")
print(textoTokenizado)

print("\nTexto após remoção de stopwords personalizadas:")
print(textoSemStopwords)
```

### **STEMMER(Derivação)**

Quando trabalhamos com textos em PLN, muitas vezes nos deparamos com palavras derivadas, ou seja, variações de uma palavra original (radical/raiz) que surgem pela adição de prefixos, sufixos ou outras transformações linguísticas. Essas modificações podem indicar tempo verbal, negação, intensidade, ou mesmo mudar a classe gramatical.

Exemplos:

feliz → infeliz (prefixo in- indica negação).

cantar → cantando / cantarei (sufixos que alteram o tempo verbal).

verdade → verdadeiro (sufixo -eiro muda a classe da palavra).

Sendo assim, o ***stemmer***(derivação) auxilia os modelos a reconhecer que palavras derivadas podem estar relacionadas semanticamente (ex.: 'correr', 'corri', 'correndo').


Para trabalharmos com ***stemmer***, utilizaremos o módulo stem da nossa velha amiga, a biblioteca *NLTK*. Mais especificamente, poderíamos usar a classe *PorterStemmer*, porém, como nem tudo são flores, a classe original possui suporte apenas para o inglês.

Por isso, no caso do nosso bom e velho português, utilizaremos o *SnowballStemmer*, também disponível no *NLTK*, que implementa regras de radicalização adequadas para a língua portuguesa.

```python
import nltk

#importando a classe SnowballStemmer da extensão snowball do módulo stem
from nltk.stem.snowball import SnowballStemmer

#instanciando a classe o stemmer para português
stemmer = SnowballStemmer("portuguese")

palavra = "correr"

#Derivando a palavra correr, ou seja, reduzindo a palavra correr ao seu radical
palavraDerivada = stemmer.stem(palavra)

#Exibindo o resultado
print( "Derivação da palavra 'Correr' = " + palavraDerivada)
```

Ops! “corr” não é uma palavra, muito menos um radical. Será que meu algoritmo está quebrado?

👉 Não nesse caso! Nosso algoritmo de ***stemming*** está funcionando corretamente. O que acontece é que, quando trabalhamos com ***stemmig*** em qualquer idioma, seja utilizando o *SnowballStemmer* ou o *PorterStemmer*, precisamos ter cuidado: o processo reduz as palavras a um radical, independentemente de esse radical ser ou não uma palavra existente.

Em outras palavras, as classes e funções que implementam a ***stemmer*** não busca correção linguística, mas sim consistência na redução. Isso costuma ser suficiente (e até desejável) em tarefas como:

- Busca textual (encontrar “correr” mesmo que o usuário digite “correndo”);

- Classificação de textos (agrupar documentos que usam variações da mesma palavra);

- Análise de frequência de termos.

### **LEMATIZAÇÃO**

Se a necessidade for reduzir as palavras a uma forma correta no idioma (ex.: “correndo” → “correr”), devemos utilizar a ***lematização***, que depende de dicionários ou modelos linguísticos.

A ***lematização*** possui o mesmo objetivo do ***stemming***: reduzir as palavras à sua forma base. A diferença é que, como utiliza ferramentas baseadas em dicionários linguísticos, o resultado será sempre uma palavra válida no idioma.

### **POS(Classificação de Classes gramaticais)**

A PoS (Part-of-Speech) é um processo fundamental em PLN. Basicamente, ela é responsável por atribuir a cada palavra do texto sua respectiva classe gramatical (substantivo, verbo, artigo, etc.), o que possibilita ao algoritmo enxergar o texto de forma estruturada. Ou seja, a PoS indica ao algoritmo a organização do texto, tornando possível uma análise integrada, permitindo o reconhecimento de padrões específicos da língua e auxiliando na desambiguação.

Exemplo -> repassamos para um algoritmo a seguinte frase para que ele classifique as palavras segundo suas respectivas classes gramaticais:

| Palavra       | Classe Gramatical (PoS)    |

| João          | Nome Próprio (Substantivo) |
| Silva         | Nome Próprio (Substantivo) |
| visitou       | Verbo (pretérito)          |
| São           | Nome Próprio (Substantivo) |
| Paulo         | Nome Próprio (Substantivo) |
| em            | Preposição                 |
| 12            | Numeral                    |
| de            | Preposição                 |
| setembro      | Substantivo (mês)          |
| de            | Preposição                 |
| 2023          | Numeral                    |
| para          | Preposição                 |
| uma           | Artigo Indefinido          |
| reunião       | Substantivo                |
| com           | Preposição                 |
| a             | Artigo Definido            |
| empresa       | Substantivo                |
| TechSolutions | Nome Próprio (Substantivo) |

### **NER(Reconhecimento de Entidade Nomeada)**

O Reconhecimento de Entidades Nomeadas, ou NER (Named Entity Recognition), é outra técnica importante no campo do Processamento de Linguagem Natural (PLN) que também auxilia na estruturação do texto, tornando o processamento pelo algoritmo mais eficiente. Seu principal objetivo é identificar e classificar automaticamente palavras ou expressões específicas dentro de um texto, chamadas de entidades nomeadas. Essas entidades geralmente incluem nomes de pessoas, locais, organizações, datas, valores monetários, entre outros.

Exemplo -> Repassamos para um algoritmo a mesma frase do conteúdo anterior, para que ele classifique as entidades nomeadas.

“O presidente João Silva visitou São Paulo em 12 de setembro de 2023 para uma reunião com a empresa TechSolutions.”

A saída esperada seria a seguinte:

| Entidade               | Categoria   |
| ---------------------- | ----------- |
| João Silva             | Pessoa      |
| São Paulo              | Local       |
| TechSolutions          | Organização |


Para começar a trabalhar com a indentificação de entidades nomeadas (NERs) presentes nos textos analisados, utilizaremos novamente a biblioteca spaCy, juntamente com o modelo *pt_core_news_sm*

Isso evidencia que, em certos contextos, apenas aplicar a PoS pode não ser suficiente, pois nós, humanos, reconhecemos naturalmente o tipo de cada substantivo próprio, mas para a máquina isso é um mistério até que as NERs sejam marcadas.

Antes da marcação das NERs, nosso algoritmo poderia interpretar que todos esses termos se referiam a locais ou empresas. No entanto, ao indicar explicitamente o que cada um representa, fornecemos ao algoritmo uma compreensão mais precisa do contexto.

Vale abrir um parêntese: as NERs não se limitam apenas a substantivos próprios — elas também podem identificar outros tipos de entidades, como datas, valores numéricos, organizações, locais, entre outros.

## **BOW (Saco de palavras)**

Um *BoW* (Bag of Words), ou saco de palavras, é uma técnica bastante utilizada em Processamento de Linguagem Natural (PLN) para representar textos numericamente.
Ela transforma um texto em um vetor, onde cada posição representa uma palavra do vocabulário, e o valor armazenado indica quantas vezes essa palavra aparece no texto.

O BoW não considera a ordem das palavras nem a gramática, apenas a frequência com que cada termo ocorre.



🧩 Exemplo 1

Frases:
 - “O gato gosta de leite.”

- “O gato gosta, gosta muito de leite!”

Vocabulário (ignorando maiúsculas e pontuação):
[“o”, “gato”, “gosta”, “de”, “leite”, “muito”]

Para transformarmos um texto em um BoW utilizaremos a biblioteca *sklearn*, mais especificamente o metodo *CountVectorizer*.

Agora, com as ferramentas necessárias, vamos definir três frases de exemplo para transformá-las em representações BoW (Bag of Words).

```python
from sklearn.feature_extraction.text import CountVectorizer

# Lista de frases a serem analisadas
frases = [
    "Eu gosto de comer chocolate",
    "Eu gosto de viajar, porém não consigo pegar um avião",
    "Será que irá chover no final de semana?"
]

# Criando o vetor de contagem de palavras
vectorizer = CountVectorizer()

# Transformando o texto em uma matriz de ocorrência de palavras
# Cada linha representa uma frase, e cada coluna representa uma palavra única
# O valor em cada célula indica quantas vezes a palavra aparece na frase
matriz_ocorrencias = vectorizer.fit_transform(frases)

# Convertendo a matriz para um array NumPy para visualização
matriz_array = matriz_ocorrencias.toarray()

# Exibindo os nomes das colunas (palavras únicas) para melhor entendimento
print("Palavras identificadas:")
print(vectorizer.get_feature_names_out())

# Exibindo a matriz de contagem
print("\nMatriz de ocorrência de palavras:")
print(matriz_array)

```

**Por que usar Bows?**
- Extração de recursos: Converte dados de texto não estruturados em uma forma estruturada (vetores), permitindo que esses dados sejam utilizados como entrada para diversos algoritmos de aprendizado de máquina.

 - Simplicidade e eficiência: O modelo BoW é computacionalmente simples de implementar e bastante eficiente, especialmente em conjuntos de dados (corpora) de texto de pequeno a médio porte.

 - Cálculo de similaridade entre documentos: Pode ser utilizado para medir a similaridade entre documentos de texto, por meio de técnicas como a similaridade do cosseno, facilitando tarefas como classificação, agrupamento e recomendação de textos.

**Limitações do modelo Bag of Words (BoW)**

 - Ignora a ordem das palavras: O BoW trata o texto como um conjunto de palavras isoladas, desconsiderando completamente a sequência em que elas aparecem. Isso significa que frases com palavras idênticas, mas em ordens diferentes, terão a mesma representação vetorial — o que pode levar à perda de informação semântica.

- Desconsidera o contexto: O modelo não capta o significado das palavras com base no seu contexto. Palavras ambíguas, como “banco” (pode ser um objeto ou uma instituição financeira), não são desambiguadas com base nas palavras vizinhas.

- Representação fixa e de comprimento igual: Todos os documentos são representados com vetores do mesmo tamanho, baseados em um vocabulário global. Essa abordagem pode ser inadequada para lidar com textos de comprimentos variados ou com estruturas muito distintas.

- Dependência de um vocabulário completo: É necessário conhecer previamente todas as palavras que podem aparecer no texto para construir o vocabulário. Em idiomas com vocabulário muito amplo ou em domínios técnicos, isso pode gerar vetores enormes e difíceis de manter.

- Sem compreensão de significado ou semântica: O BoW apenas conta a frequência das palavras, sem entender seus significados. Ele trata sinônimos como palavras distintas e não reconhece diferentes formas de uma mesma palavra (ex.: "correr", "correu", "correndo").

- Alta dimensionalidade e esparsidade: Quanto maior o vocabulário, mais esparsas (com muitos zeros) e de alta dimensão serão as representações vetoriais. Isso pode prejudicar a performance de algoritmos de aprendizado de máquina, que têm dificuldade em extrair padrões úteis de dados muito dispersos.

- Alto custo computacional com grandes conjuntos de dados: As matrizes geradas por BoW são geralmente esparsas e grandes, consumindo muita memória. Para lidar com isso, é necessário utilizar bibliotecas e estruturas otimizadas para computação esparsa, o que adiciona complexidade ao pipeline de processamento.

Algumas das limitações do modelo BoW podem ser parcialmente contornadas com o uso de recursos auxiliares ou variações na implementação. Por exemplo:

**Conversão de matrizes esparsas para densas:** A matriz gerada pelo CountVectorizer geralmente é esparsa (contém muitos zeros), o que economiza memória, mas pode dificultar operações matemáticas simples ou visualizações.
Para isso, podemos usar o método .todense() ou .toarray() para convertê-la em uma matriz densa (NumPy array). Isso facilita o manuseio da matriz, principalmente em etapas iniciais de prototipagem e análise:

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

# Textos de exemplo
textos = [
    "Gosto de programar em Python",
    "Python é uma linguagem poderosa",
    "Linguagens de programação são interessantes"
]

# Criando o vetor de contagem de palavras
vetorizar = CountVectorizer()
matriz_esparsa = vetorizar.fit_transform(textos)

print("Vocabulário:")
print(vetorizar.get_feature_names_out())

# Exibindo a matriz esparsa
print("\nMatriz esparsa:")
print(matriz_esparsa.toarray())

# Convertendo para matriz densa
matriz_densa = matriz_esparsa.todense()

print("\nMatriz densa (array completo):")
print(matriz_densa)
```

**Remoção de Stopwords:** Como já vimos anteriormente, stopwords são palavras muito frequentes em um idioma (como "o", "a", "de", "é") que ajudam na fluidez do texto, mas não carregam significado relevante para tarefas como classificação ou análise de conteúdo.

1. Ao removê-las antes de aplicar o modelo Bag of Words (BoW), conseguimos:

2. Reduzir o tamanho da matriz gerada, economizando memória e processamento;

3. Focar nos termos mais importantes do texto, que realmente indicam seu tema;

4. Evitar ruído nos dados, melhorando a performance de algoritmos de machine learning.

**Uso de n-gramas:** Para capturar parte da ordem ou contexto local, é possível usar n-gramas (ex: bigramas, trigramas), permitindo que o modelo considere combinações de palavras consecutivas.

Antes de implementarmos algo com n-gramas, primeiramente precisamos entender o que são e como “vivem”.

N-gramas são sequências de n palavras consecutivas extraídas de um texto. Eles ajudam a capturar parte da estrutura e contexto que o modelo Bag of Words tradicional ignora.

 - Um unigrama (n=1) considera palavras isoladas:
Ex: "Python", "é", "legal".

 - Um bigrama (n=2) considera pares de palavras:
Ex: "Python é", "é legal".

 - Um trigrama (n=3) considera trios de palavras:
Ex: "Python é legal".

Para extrair n-gramas usando o CountVectorizer, não é necessário mudar de ferramenta. Podemos continuar usando a mesma classe, bastando configurar o parâmetro ngram_range=(N, N) para indicar o tamanho dos n-gramas desejados.

 - ngram_range=(1, 1) → unigramas (padrão)

 - ngram_range=(2, 2) → bigramas

 - ngram_range=(3, 3) → trigramas

```python
from sklearn.feature_extraction.text import CountVectorizer

texto = [
    "Aprender Python é divertido",
    "Python é uma linguagem poderosa"
]

# Extraindo unigramas e bigramas
vectorizer = CountVectorizer(ngram_range=(1, 3))
# Para testar com apenas bigramas basta mudar o parametro para(2, 2)
# Para testar com apenas trigrmas basta mudar o parametro para(3, 3)
# Para testar com unigramas, bigramas e trigramas basta mudar o parametro para(1, 3)
matriz = vectorizer.fit_transform(texto)

print("Bigramas encontrados:")
print(vectorizer.get_feature_names_out())

print("\nMatriz de frequência:")
print(matriz.toarray())
```

## **TF-IDF (Term Frequency - Inverse Document Frequency)**

TF-IDF (Frequência de Termos - Frequência Inversa de Documentos) é uma maneira de medir a relevância de uma palavra (atribuindo-lhe uma pontuação) para um documento em uma coleção de documentos. Diferente dos demais conceitos que vimos anteriormente, o TF-IDF é uma heurística, ou seja, um conjunto de regras e conceitos que podem ser interpretados e construídos de formas diferentes, mas que seguem os mesmos paradigmas.

O TF-IDF busca mapear a Frequência de Termos – Frequência Inversa de Documentos, que é uma estatística numérica que visa refletir a importância de uma palavra para um documento dentro de uma coleção ou corpus(conjunto de textos). Nesse caso, o TF analisa a frequência com que uma palavra aparece em um determinado texto, enquanto o IDF analisa, em um contexto mais amplo, a frequência dessas palavras no conjunto de textos analisados.

Juntando os resultados de ambos os processos, obtemos um vetor com valores numéricos que representam pesos para as palavras. Quanto maior o peso, mais importante aquela palavra é para o que está sendo analisado.

Uma das formas mais comuns de se calcular o **TF (Term Frequency)** é dividindo a frequência de uma palavra no texto pelo **número total de termos** desse texto.

Já o **IDF (Inverse Document Frequency)** é calculado por meio do **logaritmo da razão entre o número total de textos e o número de textos em que um determinado termo aparece**.

Essas duas medidas são usadas em conjunto (TF-IDF) para identificar a **importância de um termo em um conjunto de documentos**.


Fórmulas

**TF(t) = (número de vezes que o termo t aparece no texto) / (número total de termos do texto)**
**IDF(t) = log10(N / nt)**

onde:

* **N** = número total de textos do conjunto analisado;
* **nt** = número de textos em que o termo t aparece.

```python
#Importando ferramentas para o calculo de TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
```

Posteriormente, definimos os textos de exemplo que serão passados como parâmetro para a função. Em seguida, instanciamos a classe *TfidfVectorizer* e passamos como parâmetro o corpus que desejamos processar. Após isso, imprimimos o resultado obtido.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Definindo um corpus de exemplo
corpus = [
    "Meu cachorro brinca muito bem de buscar",
    "Eu tinha um cachorro que não parecia um cachorro, mas era um cachorro",
    "cachorro coelho gato coelho galo cachorro porco cabra cavalo cachorro cachorro gato vaca leão marinho pássaro pombo pinguins e baleias e baús de tesouro e que outros itens aleatórios podem entrar neste longo documento, eu me pergunto",
    "Eu tenho um gato",
    "Eu tenho uma cabra",
    "Por que todo mundo está falando sobre seus animais de estimação",
    "Eu tenho uma zebra",
    "Você não tem uma zebra, Gustavo"
]

# Instanciando a classe TfidfVectorizer
tfidf = TfidfVectorizer()

# Aplicando o fit_transform para gerar a matriz TF-IDF
response = tfidf.fit_transform(corpus)

# Obtendo os nomes das features (palavras)
feature_names = tfidf.get_feature_names_out()

# Imprimindo o vocabulário e seus respectivos valores TF-IDF para o primeiro documento
for col in response.nonzero()[1]:
    print(feature_names[col], '-', response[0, col])
```

 - Motores de busca (ex.: Google): O TF-IDF ajuda a ranquear os resultados de pesquisa, atribuindo maior relevância às páginas que contêm os termos mais importantes relacionados à busca do usuário.

- Análise de similaridade entre documentos: É usado para medir o grau de semelhança entre textos, auxiliando na detecção de plágio ou na identificação de conteúdos duplicados.

É importante lembrar que o TF-IDF é uma heurística e pode ser encontrado em diversas formulações, como o Okapi BM25. Não existe uma versão universalmente superior, pois a eficácia de cada abordagem depende do contexto e do tipo de aplicação.

No entanto, todas essas variações compartilham um princípio fundamental: palavras raras devem receber maior peso ou importância do que palavras muito frequentes, pois tendem a ser mais informativas para a distinção entre documentos.

## **EMBEDDINGS**

Em PLN, *embeddings* são representações vetoriais multidimensionais de palavras, frases ou textos.

Mas o que isso significa?
Máquinas não conseguem compreender nada além de representações numéricas. No entanto, com o avanço da computação e a crescente necessidade de lidar com outros tipos de dados, foram desenvolvidos diversos processos e algoritmos com a finalidade de converter informações linguísticas em números, que são o tipo de dado que os computadores conseguem manipular e interpretar.

Dentro desse contexto, surgiram e evoluíram diferentes métodos e técnicas de representação semântica, sendo um dos mais importantes os *embeddings*. Em suma, *embeddings* são vetores numéricos que representam palavras, frases ou textos em um espaço multidimensional, onde cada dimensão(ou posição do vetor) captura algum aspecto semântico ou sintático do conteúdo.

Por exemplo, palavras com significados semelhantes (como “rei” e “rainha”) tendem a ter vetores próximos(valores semelhantes) nesse espaço, enquanto palavras com significados distintos (como “rei” e “carro”) ficam mais distantes. Essa capacidade de preservar relações de similaridade e contexto faz dos *embeddings* uma das principais bases para o funcionamento de modelos modernos de linguagem, como o *Word2Vec, GloVe, BERT e GPT.*

Para transformar textos em *embeddings*, são utilizadas [redes neurais](https://www.ibm.com/br-pt/think/topics/neural-networks), que aprendem a associar cada palavra, frase ou texto a uma representação vetorial. O processo, de forma simplificada, pode ser descrito assim:

entrada: "carro"

→ processamento pela rede neural

→ saída: [-0.12, 0.34, 0.54, ...]

Nesse exemplo, a palavra "carro" é convertida em um vetor de números reais, geralmente com dezenas ou centenas de dimensões.
Cada número desse vetor representa uma característica aprendida pela rede — por exemplo, aspectos semânticos (como o fato de “carro” estar relacionado a “motor”, “estrada”, “transporte”) ou sintáticos (como a categoria gramatical da palavra).

Essas redes são treinadas em grandes volumes de texto e aprendem a reconhecer padrões de coocorrência, ou seja, a forma como as palavras aparecem juntas em contextos semelhantes.
Assim, palavras que ocorrem em contextos parecidos acabam tendo vetores próximos no espaço multidimensional.

### Embeddings independentes de contexto

Com modelos de incorporação independentes de contexto, as representações aprendidas são fixas e únicas para cada palavra, sem levar em consideração o ambiente em que ela aparece. Isso implica que homônimos, como “banco” (que pode se referir a uma instituição financeira ou a um assento), dificilmente terão seus significados desambiguados pelo contexto. Como resultado, a palavra receberá praticamente o mesmo vetor em situações completamente diferentes, mesmo que represente conceitos distintos.

Por exemplo, nas frases:

 - “Eu fui ao banco sacar dinheiro”

 - “Eu me sentei no banco e olhei a praça”

A palavra “banco” provavelmente terá uma representação vetorial quase idêntica em ambos os casos, apesar de assumir significados totalmente diferentes em cada contexto.

Para trabalhar com incorporações independentes de contexto utilizamos o modelo *Word2Vec*, especificamente com a abordagem *Skip-gram*. Esse método aprende representações vetoriais ao treinar a rede para prever palavras que aparecem ao redor de uma palavra central em uma frase.

Como funciona:

 - **Seleção da palavra central**: O modelo escolhe uma palavra específica em uma sentença.

 - **Definição da janela de contexto**: Determina-se um intervalo (por exemplo, 5 palavras antes e 5 depois) que representa o entorno da palavra central.

 - **Previsão das palavras do contexto**: O modelo é treinado para prever as palavras que aparecem dentro dessa janela de contexto, usando apenas a palavra central como entrada.

Para trabalharmos com o *Word2Vec*, primeiramente iremos instalar as dependências necessárias:

```python
# Instalar a biblioteca (caso ainda não tenha)
!pip install gensim
```

```python
# Importar o Word2Vec(Modelo pré-treinado que irá gerar os embeddings)
from gensim.models import Word2Vec

# Criar um pequeno corpus de exemplo (nesse caso, uma lista de frases já tokenizadas)
sentences = [
    ["rei", "governa", "um", "reino"],
    ["rainha", "governa", "um", "reino"],
    ["homem", "é", "forte"],
    ["mulher", "é", "sábia"],
    ["o", "rei", "e", "a", "rainha"],
    ["o", "homem", "e", "a", "mulher"]
]

# Parametrizamos nosso modelo que irá gerar os embeddings)
model = Word2Vec(
    sentences,     # dados de treino
    vector_size=100,  # 100 dimensões para cada palavra
    window=5,      # olha até 5 palavras antes/depois
    min_count=2,   # ignora palavras raras
    sg=1,          # usa Skip-gram
    epochs=10,     # treina por 10 épocas
    workers=4      # usa 4 núcleos do processador
)


# Obter o embedding da palavra "rei"
embedding_rei = model.wv["rei"]
print("Embedding para 'rei':\n", embedding_rei)

# Mostrar as palavras mais semelhantes a "rei"
similar = model.wv.most_similar("rei", topn=3)
print("\nPalavras mais semelhantes a 'rei':")
for palavra, similaridade in similar:
    print(f"{palavra}: {similaridade:.3f}")
```

Ou seja, nesse caso, o foco está em representar a palavra a partir das palavras que a cercam no contexto, e não apenas em seu significado isolado ou no significado individual das palavras ao redor.

### Embeddings dependentes de contexto

Ao contrário das representações vetoriais de palavras independentes de contexto, os métodos dependentes de contexto aprendem vetores que muito provavelmente serão diferentes, mesmo quando a palavra é a mesma. Isso ocorre porque, para gerar o embedding, esses modelos não consideram apenas a palavra isolada na frase, mas também o contexto que ela estabelece dentro dela. Em modelos mais recentes, avalia-se ainda  quanto outras palavras influenciam a palavra em análise, ajustando sua representação de acordo com o contexto específico.”

Nesse caso, no exemplo:

 - “Eu fui ao banco sacar dinheiro”
 - “Eu sentei no banco e olhei a praça”,

A palavra "banco", tera representações vetoriais distintas, calculdas com base no todo da frase.

Exemplo prático em Python com *BERT*:

```python
# Instalar a biblioteca (caso ainda não tenha)
!pip install transformers torch --quiet
```

```python

#Importar o BertModel
from transformers import BertTokenizer, BertModel
import torch

# Carrega o modelo pré-treinado BERT base já baixada
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Duas frases com a palavra "bank" em contextos diferentes
sent1 = "I sat on the bank of the river."
sent2 = "I deposited money in the bank."

# 1. Tokenizar e converter para tensores
inputs1 = tokenizer(sent1, return_tensors='pt')
inputs2 = tokenizer(sent2, return_tensors='pt')

# 2. Obter embeddings
with torch.no_grad():
    outputs1 = model(**inputs1)
    outputs2 = model(**inputs2)

# Pegar o embedding da palavra "bank" em cada sentença
# (posição depende da tokenização)
index1 = (inputs1['input_ids'][0] == tokenizer.convert_tokens_to_ids('bank')).nonzero(as_tuple=True)[0]
index2 = (inputs2['input_ids'][0] == tokenizer.convert_tokens_to_ids('bank')).nonzero(as_tuple=True)[0]

embedding1 = outputs1.last_hidden_state[0][index1]
embedding2 = outputs2.last_hidden_state[0][index2]

# Calcular similaridade de cosseno
cos = torch.nn.functional.cosine_similarity(embedding1, embedding2)
print(f"Similaridade entre 'bank' (sentenças diferentes): {cos.item():.3f}")
```

### Embeddings de sentenças

Os modelos de *embeddings* tradicionais, como os baseados em *Word2Vec* ou *GloVe*, normalmente geram representações vetoriais apenas para palavras individuais. Como consequência, para processar frases ou textos completos, tornou-se necessário desenvolver heurísticas que permitissem combinar esses vetores em uma única representação significativa.

Diversas estratégias foram criadas para capturar o sentido global de uma sentença, parágrafo ou documento. Entre as abordagens mais modernas está o uso do token especial *CLS*, presente em arquiteturas como *BERT*. Esse token recebe, após o processamento do texto pelo modelo, uma representação vetorial que funciona como um resumo semântico da sequência inteira.

Antes de modelos como o *BERT*, uma estratégia comum era a agregação de embeddings. Nesse método, cada palavra da frase é convertida em seu respectivo vetor, e então aplica-se algum tipo de combinação, como média, soma ponderada ou concatenação, para gerar um único vetor representativo.

Por exemplo, na frase:

“Eu gosto de chocolate”,

Nesse caso, cada palavra é convertida em seu embedding correspondente. Em seguida, esses vetores são agregados — frequentemente por meio da média vetorial — resultando em um embedding único que busca representar o significado geral da frase.

Como o *CLS* é um artefato gerado por arquiteturas do tipo *BERT*, e como no tópico anterior será utilizada essa mesma arquitetura, para instalar os requisitos necessários basta:

```python
# Instalar a biblioteca (caso ainda não tenha)
!pip install transformers torch --quiet
```

```python
from sentence_transformers import SentenceTransformer
import torch

# Carrega o modelo
model = SentenceTransformer("all-MiniLM-L6-v2")

# Frase desejada
sentence = "Eu gosto de café."

# Tokenização
encoded = model.tokenize([sentence])

# Passa pelo modelo (gera embeddings por token)
output = model(encoded)

# Todos os embeddings de tokens
token_embeddings = output["token_embeddings"]  # shape: (1, seq_len, 384)


# CLS = primeiro token da sequência
cls_embedding = token_embeddings[0, 0, :]   # shape: (384,)

print("Embedding CLS da frase:")
print(cls_embedding)
print("Dimensão:", cls_embedding.shape)
```

No entanto, existem outras técnicas utilizadas para a conversão de sentenças em embeddings, por exemplo:

 **1. *Mean Pooling* (média dos embeddings dos tokens):** Nesse método, cada palavra da sentença recebe um embedding próprio (um vetor numérico).
Para gerar o embedding final da frase, tiramos a **média** desses vetores.

Exemplo:

Imagine que sua frase tenha 3 palavras, e o modelo gere os seguintes embeddings simplificados com tres dimensões:

palavra1 → [1.0, 0.2, 0.4]
palavra2 → [0.7, 0.1, 0.5]
palavra3 → [0.9, 0.3, 0.3]

| Palavra  | Dim 1 | Dim 2 | Dim 3 |
| -------- | ----- | ----- | ----- |
| palavra1 | 1.0   | 0.2   | 0.4   |
| palavra2 | 0.7   | 0.1   | 0.5   |
| palavra3 | 0.9   | 0.3   | 0.3   |


-> Para fazer **Mean Pooling**, calculamos a média coluna a coluna:

| Dimensão | Cálculo               | Média    |
| -------- | --------------------- | -------- |
| Dim 1    | (1.0 + 0.7 + 0.9) / 3 | **0.87** |
| Dim 2    | (0.2 + 0.1 + 0.3) / 3 | **0.20** |
| Dim 3    | (0.4 + 0.5 + 0.3) / 3 | **0.40** |


-> O resultado final seria um unico embedding, que represnetaria os tres:

Embedding da frase → [0.87, 0.20, 0.40]





***2. Max Pooling***  **(máximo elemento por dimensão):** Aqui, em vez de calcular a média, escolhemos o maior valor de cada dimensão dos vetores das palavras.

-> Usando os mesmos embeddings do exemplo anterior:

palavra1 → [1.0, 0.2, 0.4]
palavra2 → [0.7, 0.1, 0.5]
palavra3 → [0.9, 0.3, 0.3]

| Palavra  | Dim 1 | Dim 2 | Dim 3 |
| -------- | ----- | ----- | ----- |
| palavra1 | 1.0   | 0.2   | 0.4   |
| palavra2 | 0.7   | 0.1   | 0.5   |
| palavra3 | 0.9   | 0.3   | 0.3   |

-> Aplicando **Max Pooling**, que analisa na mesma posição qual é o malor mais sobressalente :

| Dimensão | Valores       | Máximo  |
| -------- | ------------- | ------- |
| Dim 1    | 1.0, 0.7, 0.9 | **1.0** |
| Dim 2    | 0.2, 0.1, 0.3 | **0.3** |
| Dim 3    | 0.4, 0.5, 0.3 | **0.5** |

-> O embedding final destaca as características mais fortes entre todas as palavras da sentença.

Embedding da frase → [1.0, 0.3, 0.5]





### Aplicações dos Embeddings

Os *embeddings* são fundamentais para uma ampla gama de tarefas em PLN e IA:

 - Busca por Similaridade Semântica: Em vez de buscar por correspondência exata de palavras-chave, os embeddings permitem buscar documentos ou trechos de texto com significado semelhante. Isso é usado em motores de busca para fornecer resultados mais relevantes.

 - Sistemas de Recuperação e Geração Aumentada (*RAG*): São a base para sistemas de *RAG*, onde informações relevantes são recuperadas de um vasto conjunto de dados e usadas para enriquecer as respostas de modelos de linguagem (*LLMs*).

 - Classificação de Texto: Podem ser usados para categorizar textos automaticamente, como classificar e-mails como spam/não spam, ou notícias por tópico.

 - Agrupamento (*Clustering*): Permitem agrupar textos semanticamente semelhantes sem a necessidade de rótulos pré-definidos.

 - Tradução e Resumo: A representação vetorial do significado subjacente de um texto facilita tarefas complexas como tradução automática e sumarização de documentos.

 - Análise de Sentimento: Ajudam a analisar o tom emocional de um texto, identificando se é positivo, negativo ou neutro.

 - Redução de Dimensionalidade: Transformam representações de texto esparsas e de alta dimensão em vetores densos de dimensão muito menor, o que torna o processamento mais rápido e eficiente computacionalmente.

## **APRENDIZAGEM DE MÁQUINA**

**O Que é Aprendizado de Máquina?**

O Aprendizado de Máquina (do inglês *Machine Learning*) é uma das ferramentas fundamentais na construção de sistemas de Inteligência Artificial (IA). Trata-se de um campo que reúne técnicas algorítmicas capazes de criar modelos (como redes neurais ou árvores de decisão) que aprendem a resolver problemas de forma autônoma.

Ao contrário da programação tradicional, onde as regras são escritas manualmente, no *Machine Learning* o sistema utiliza grandes volumes de dados para "aprender" as regras por conta própria. Para isso, o processo é dividido fundamentalmente em duas etapas de dados:

Dados de Treinamento: Usados para ajustar os parâmetros do modelo. É aqui que o algoritmo "estuda" os padrões.

Dados de Teste: Usados para avaliar o desempenho do modelo em dados inéditos. O objetivo é garantir a generalização, ou seja, que o modelo funcione bem no mundo real e não apenas decore os dados de treino.


**Fluxo geral de um projeto de Machine Learning:**

1.	Definição do problema: classificação? regressão? detecção de anomalias?
2.	Coleta de dados: reunir exemplos relevantes (rótulos para aprendizagem supervisionada).
3.	Pré-processamento: limpeza, tratamento de valores faltantes, normalização, codificação de categorias.
4.	Divisão dos dados: normalmente em treino / validação / teste (ex.: 70/15/15).
o	Treino: ajustar parâmetros.
o	Validação: escolher hiperparâmetros e evitar overfitting.
o	Teste: avaliação final, nunca usado durante ajuste.
5.	Treinamento do modelo: executar o algoritmo sobre os dados de treino.
6.	Avaliação: usar métricas apropriadas (precisão, recall, F1, AUC, RMSE, etc.).
7.	Ajuste e validação cruzada: por exemplo k-fold cross-validation.
8.	Implantação e monitoramento em produção (dados mudam com o tempo — data drift).


Exemplos:


1. Árvores de Decisão (*Decision Trees*)
Este modelo funciona como um fluxograma de "sim" ou "não", dividindo os dados em ramificações até chegar a uma conclusão. É muito intuitivo e fácil de interpretar.

Exemplo Prático (Aprovação de Empréstimo): O banco quer saber se deve emprestar dinheiro a um cliente. A árvore faz perguntas sequenciais:

O cliente tem o nome limpo? (Se não → Nega).

A renda é superior a R$ 3.000? (Se sim → Aprova). O caminho percorrido nas "folhas" da árvore determina o resultado final.

2. *Random Forests* (Florestas Aleatórias)
Como o nome sugere, este algoritmo cria uma "floresta" com diversas Árvores de Decisão diferentes e combina os resultados de todas elas para dar uma resposta mais robusta e precisa. Isso evita erros que uma única árvore poderia cometer sozinha.

Exemplo Prático (Preço de Imóveis): Para estimar o preço de uma casa, uma árvore pode focar muito na localização, enquanto outra foca no tamanho. A Random Forest pega a opinião de centenas dessas árvores e calcula a média, resultando em uma avaliação de preço muito mais assertiva do que a de um único corretor.

3. SVM (*Support Vector Machine*s)
O SVM é um algoritmo poderoso para classificação. Ele tenta encontrar a melhor linha (ou hiperplano, em dimensões superiores) que separa dois grupos de dados com a maior margem de segurança possível.

Exemplo Prático (Classificação de E-mail): O filtro de spam do seu e-mail pode usar SVM. Ele analisa características (como palavras-chave, remetente, links) e traça uma linha separadora: de um lado fica o que é "Spam", do outro o que é "Caixa de Entrada". O SVM busca a separação mais clara possível para não jogar um e-mail importante no lixo eletrônico.

Nesta etapa, o dataset de análise de sentimentos é baixado automaticamente do *Kaggle* por meio da biblioteca *kagglehub*. Após o download, o arquivo *CSV* presente no diretório é identificado e carregado em um *DataFrame* do *pandas*. Essa estrutura facilita a manipulação dos dados e o acesso às colunas de texto e rótulos de sentimento.

```python
# Baixa o dataset de análise de sentimentos de redes sociais do Kaggle
path = kagglehub.dataset_download(
    "kashishparmar02/social-media-sentiments-analysis-dataset"
)

# Lista todos os arquivos CSV existentes no diretório do dataset
csv_files = [f for f in os.listdir(path) if f.endswith(".csv")]
```

Com o *dataset* carregado, são selecionadas as colunas relevantes para o problema. A coluna que contém os textos representa as entradas do modelo, enquanto a coluna de sentimentos corresponde aos rótulos que o classificador deverá aprender a prever. Os textos e rótulos são convertidos para listas, garantindo compatibilidade com as etapas seguintes do processamento.

```python
# Seleciona o primeiro arquivo CSV encontrado
file_path = os.path.join(path, csv_files[0])

# Lê o arquivo CSV e carrega os dados em um DataFrame
df = pd.read_csv(file_path)


# Nome da coluna que contém os textos
text_col = "Text"

# Nome da coluna que contém os rótulos (sentimentos)
label_col = "Sentiment"

# Converte os textos para string e cria uma lista
texts = df[text_col].astype(str).tolist()

# Converte os rótulos para string e cria uma lista
labels = df[label_col].astype(str).tolist()
```

Em seguida, é carregado o tokenizador e o modelo *BERT* pré-treinado *(bert-base-uncased)*. Nesse projeto, o *BERT* não é treinado novamente; ele é utilizado apenas como um extrator de características, responsável por transformar textos em embeddings numéricos que capturam informações semânticas e contextuais.

```python
# Carrega o tokenizador do modelo BERT base (uncased = sem diferenciar maiúsculas/minúsculas)
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

# Carrega o modelo BERT pré-treinado
# OBS: aqui estamos usando o BERT apenas como extrator de características (embeddings)
model = BertModel.from_pretrained("bert-base-uncased")
```

Os textos selecionados são tokenizados utilizando o tokenizador do *BERT*. Nessa etapa, os textos são convertidos em tokens compreendidos pelo modelo, com aplicação de padding para igualar o tamanho das sequências(*padding*) e *truncation* para limitar textos muito longos. Para reduzir o custo computacional, apenas um subconjunto de 52 amostras é utilizado.

```python
# Tokeniza apenas as 52 primeiras amostras (provavelmente para reduzir custo computacional)
inputs = tokenizer(
    texts[:52],              # textos de entrada
    padding=True,             # adiciona padding para igualar o tamanho das sequências
    truncation=True,          # corta textos muito longos
    return_tensors="pt"       # retorna tensores PyTorch
)


# Desativa o cálculo de gradientes (não vamos treinar o BERT)
with torch.no_grad():
    outputs = model(**inputs)
```

Com os textos tokenizados, o modelo *BERT* é executado em modo de inferência, sem cálculo de gradientes. O vetor correspondente ao token especial *CLS* é extraído para cada texto, pois ele representa semanticamente a sentença como um todo. O resultado é um conjunto de vetores de 768 dimensões, que servem como representação numérica dos textos.

```python
# outputs.last_hidden_state tem dimensão:
# (n_amostras, n_tokens, 768)

# Selecionamos o embedding do token [CLS] (posição 0),
# que representa semanticamente a frase inteira
cls_embeddings = outputs.last_hidden_state[:, 0, :].numpy()

# Seleciona os rótulos correspondentes às 52 amostras usadas
labels = np.array(labels[:52])
```

Os *embeddings* e seus respectivos rótulos são divididos em conjuntos de treino e teste. Essa separação permite avaliar a capacidade de generalização do modelo, sendo utilizados 80% dos dados para treino e 20% para teste.

```python
# Divide os dados em:
# 80% treino e 20% teste
X_train, X_test, y_train, y_test = train_test_split(
    cls_embeddings,
    labels,
    test_size=0.2,
    random_state=42
)
```

Com os dados preparados, é criado um classificador *SVM* com *kernel linear*. Esse modelo é treinado utilizando os embeddings gerados pelo *BERT* como entrada. O *SVM* aprende a separar os vetores de acordo com as classes de sentimento presentes no dataset.

```python
# Cria um classificador SVM com kernel linear
svm_model = SVC(kernel="linear")

# Treina o SVM usando os embeddings do BERT
svm_model.fit(X_train, y_train)
```

Após o treinamento, o modelo é utilizado para realizar previsões sobre o conjunto de teste. O desempenho é avaliado por meio da acurácia, permitindo uma análise mais detalhada da qualidade da classificação. Por fim, aplica-se a técnica de Análise de Componentes Principais (*PCA*) para reduzir a dimensionalidade dos embeddings de 768 para 2 dimensões. Essa redução é utilizada apenas para visualização, possibilitando observar graficamente a distribuição das amostras e a separação entre as classes de sentimento no espaço vetorial.

```python
# Realiza as previsões no conjunto de teste
y_pred = svm_model.predict(X_test)

# Reduz os embeddings (768 dimensões) para apenas 2 dimensões
# para permitir visualização em gráfico
pca = PCA(n_components=2)

# Aplica o PCA apenas nos dados de treino
X_pca = pca.fit_transform(X_train)

# Cria a figura do gráfico
plt.figure(figsize=(8,6))

# Obtém os rótulos únicos (ex: positivo, negativo, neutro)
unique_labels = np.unique(y_train)

# Plota cada classe com uma cor diferente
for label in unique_labels:
    mask = y_train == label
    plt.scatter(
        X_pca[mask, 0],
        X_pca[mask, 1],
        label=label,
        s=30
    )

# Configuração do gráfico
plt.title("Distribuição das amostras (PCA dos embeddings)")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.legend(title="Sentimento")
plt.show()

# Calcula e exibe a acurácia
print("Acurácia:", accuracy_score(y_test, y_pred))

# Exibe métricas detalhadas: precisão, recall e F1-score
print("\nRelatório de Classificação:\n")
print(classification_report(y_test, y_pred))
```

Este tópico foi desenvolvido com o objetivo de ilustrar, de forma conceitual, o pipeline adotado por sistemas que realizam *Natural Language Understanding (NLU)* e *Natural Language Generation (NLG)*, como os utilizados por ferramentas do Google e pelo ChatGPT. ### **NLU**

O *Natural Language Understanding (NLU),* ou, em bom e velho português, Compreensão de Linguagem Natural, nada mais é do que o processo — ou *pipeline* — utilizado pelas máquinas para processar uma entrada de linguagem natural.
Vamos trabalhar com um exemplo textual para facilitar a compreensão.

Para entender como o *Natural Language Understanding* processa a entrada, considere a frase:

"Um novo celular será lançado no próximo ano."

1. **Pré-processamento de texto** : O primeiro passo é limpar e normalizar a entrada. Isso envolve dividir a frase em palavras individuais (tokenização), remover stopwords comuns e também reduzir as palavras às suas raízes (como converter "launched" para "launch"). Isso resulta em uma representação simplificada e mais significativa da frase.
Saída: Nesta etapa, elementos não essenciais são removidos e o texto é transformado em uma lista básica de palavras significativas.

2. **Marcação de Classes gramaticais** (POS) : Cada palavra recebe uma categoria gramatical, como substantivo, verbo ou adjetivo. Isso ajuda a identificar a função de cada palavra na frase.
Resultado: A marcação POS ajuda o sistema a entender quais palavras servem como sujeitos, ações ou descritores, contribuindo para a compreensão da estrutura da frase.

3. **Reconhecimento de Entidades Nomeadas** (NER) : Nesta fase, tipos específicos de informações, como nomes de produtos, datas ou locais, são identificados. Na frase de exemplo, "celular" pode ser reconhecido como um produto e "próximo ano" como uma referência temporal.
Resultado: O NER destaca as partes mais informativas da frase, permitindo que o sistema entenda o que e quando algo está sendo discutido.

4. **Análise de Dependências** : A análise de dependências examina como as palavras estão conectadas. Ela identifica quais palavras dependem umas das outras para transmitir significado. Por exemplo, "celular" depende de "lançado" e "próximo ano" fornece uma referência temporal para essa ação.
Saída: Esta análise mostra as relações entre as palavras, permitindo que o sistema interprete como a frase está estruturada semanticamente.

5. **Ambiguidade no sentido das palavras** : Algumas palavras podem ter múltiplos significados dependendo do contexto. Aqui, a palavra "celular" pode se referir a um telefone ou a um objeto em movimento. Ao verificar as palavras ao redor, como "lançado" e "ano", o sistema mostra que a frase se refere ao lançamento de um produto, especificamente um smartphone.
Resultado: Esta etapa garante que as palavras sejam interpretadas corretamente com base em seu uso no contexto.

6. **Reconhecimento de Intenção** : O reconhecimento de intenção identifica o propósito por trás da entrada. Nesse caso, a frase provavelmente se destina a informar sobre o lançamento de um produto. Determinar a intenção é particularmente importante em sistemas de diálogo, onde a compreensão dos objetivos do usuário é essencial.
Saída: O sistema categoriza a entrada sob uma intenção, como inform_product_release, orientando ações ou respostas apropriadas.

7. **Geração de Saída** : Uma vez compreendida a frase, o sistema formula uma resposta ou ação adequada. Por exemplo, pode responder com uma confirmação ou solicitar mais detalhes, dependendo do contexto da conversa.
Saída: Uma resposta é produzida com base no significado extraído e na intenção reconhecida, o que ajuda a manter uma interação significativa.

### **NLG**

Já o *Natural Language Generation (NLG)*, ou Geração de Linguagem Natural, é o campo do PLN que se preocupa em otimizar a criação de textos ou fala em assistentes, como a Alexa, partindo de uma base de dados estruturada em linguagem de máquina.

Pode até parecer simples, mas é importante lembrar que a linguagem natural nem sempre é composta apenas por termos técnicos ou por frases desconexas provenientes de diferentes bases. Ela precisa apresentar fluidez e coerência, o que pode representar um grande desafio para algoritmos e máquinas.

De forma abreviada e generalizada, as etapas que uma máquina, ou algoritimo seguem para a geração de conteudos em linguagem natural seria:

1. **Determinação de Conteúdo:** O sistema decide quais informações dos dados de entrada são relevantes e devem ser mencionadas. Isso envolve a filtragem de fatos com base no contexto ou na importância.
2. **Estruturação do Documento:** O conteúdo é organizado em uma estrutura contínua. São tomadas decisões sobre a ordem em que os tópicos ou fatos devem ser apresentados.
3. **Agregação:** os fatos são agrupados para melhorar a fluência e reduzir a redundância. Isso garante que o texto seja lido naturalmente, como se um ser humano resumisse vários pontos de dados.
4. **Lexicalização:** Palavras e expressões apropriadas são escolhidas para representar os fatos.
5. **Geração de expressões de referência:** o sistema gera referências a entidades como “ele”, “eles” ou nomes próprios para manter clareza e consistência em todo o texto.
6. **Realização Linguística:** Regras gramaticais são aplicadas para construir frases bem formadas.

```python
from transformers import BertTokenizer, BertModel
import torch
import numpy

tokenizer  = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")

sent1 = "I sat on the bank of the river."
sent2 = "I deposited money in the bank."

input = tokenizer(
    [sent1, sent2],
    padding=True,
    truncation=True,
    return_tensors="pt"
)

with torch.no_grad():
    output = model(**input)

embedding1 = output.last_hidden_state[0, 0, :]
embedding2 = output.last_hidden_state[1, 0, :]

cos = torch.nn.functional.cosine_similarity(
    embedding1,
    embedding2,
    dim=0
)
print(f"Similaridade : {cos.item():.3f}")
```
